```python
from snorkel import SnorkelSession
from snorkel.parser import CorpusParser
from snorkel.models import Document, Sentence, Candidate
import numpy as np
import pandas as pd
import tqdm
```

```python
session = SnorkelSession()
```

```python
sents = session.query(Sentence).all()
len(sents)
```

```python
def get_ent_frequency():
    res = []
    for sent in tqdm.tqdm(sents):
        types = list(sent.entity_types)
        words = list(sent.words)
        nwrd, ntyp = None, None
        for i, typ in enumerate(types):
            if typ == 'O':
                if ntyp:
                    res.append((ntyp, ' '.join(nwrd)))
                ntyp, nwrd = None, None
                continue
            ntyp = typ
            nwrd = (nwrd or []) + [words[i]]
        if ntyp:
            res.append((ntyp, ' '.join(nwrd)))
    return pd.DataFrame(res, columns=['typ', 'token'])
cts = get_ent_frequency()
```

```python
pd.set_option('display.max_rows', 1000)
cts.groupby(['typ', 'token']).size().rename('count').reset_index()\
    .groupby('typ', group_keys=False).apply(lambda g: g.sort_values('count', ascending=False).head(100))\
    .pipe(lambda df: pd.concat([g.reset_index(drop=True) for _, g in df.groupby('typ')], axis=1))
```
